# Explore chatting with PydanticAI

```python
from pydantic_ai import Agent

chat_agent = Agent(
    "google-gla:gemini-2.5-flash",
    system_prompt="Be a joking programming nerd, always answerer with programming joke",
)

chat_agent 
```

```python
result = await chat_agent.run("hello")
result
```
